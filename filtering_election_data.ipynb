```python
import pandas, numpy
from matplotlib import pyplot
```

```python
elections = pandas.read_csv('countypres_2000-2016.csv')
```

```python
elections = elections[elections['party'] != 'green']
```

```python
elections = elections[elections['party'] != 'NaN']
```

```python
elections = elections[pandas.notnull(elections['party'])]
```

```python
results_2016 = elections[elections['year']== 2016]
```

```python
results_2016 = results_2016.drop(['version','FIPS',"office"], axis=1)
```

```python
gop = results_2016[results_2016['party']=='republican']
dems = results_2016[results_2016['party']=='democrat']
    
```

```python
gop
```

```python
gop['percent_republican'] = (gop['candidatevotes']/gop['totalvotes'])*100
```

```python
gop
```

```python
dems['percent_democrat'] = (dems['candidatevotes']/dems['totalvotes'])*100
```

```python
gop['percent_democrat'] = dems['percent_democrat']
```

```python
new_table = pandas.merge(gop, dems, how='inner', on='county')
```

```python
new_table = new_table.drop(['year_y','state_y','state_po_y','candidate_x','candidate_y','party_x','party_y'],axis='columns')
```

```python
new_table = new_table.drop(['percent_democrat_x','totalvotes_y'],axis='columns')
```

```python
new_table = new_table.rename(columns = {'year_x':'year', 'state_x':'state','state_po_x':'state_abbr','candidatevotes_x':'r_candidate_votes'})
```

```python
new_table = new_table.rename(columns = {'totalvotes_x':'total_votes','candidatevotes_y':'d_candidate_votes','percent_democrat_y':'percent_democrat'})
```
