```python
import pandas
```

```python
by_county = pandas.read_csv('Final_election_results2016_county.csv')
```

```python
by_county = by_county.drop(['percent_republican','percent_democrat','%difference(d_favor)'],axis='columns')
```

```python
by_county
```

```python
by_state = by_county.groupby(['state_abbr']).agg({'D_candidate_votes':"sum",'R_candidate_votes':'sum','totalvotes':'sum'})
```

```python
by_state['%rep'] = (by_state['R_candidate_votes']/by_state['totalvotes'])* 100
by_state['%dem'] = (by_state['D_candidate_votes']/by_state['totalvotes'])* 100
by_state['%diff_dem_favor'] = by_state['%dem']
```
